# Matrix Multiplication with @ and .matmul()

<img src=https://i.imgur.com/LCdhmi8.png width=350>

```python
a1 = np.array([[1, 3],
               [0, 1],
               [6, 2],
               [9, 7]])

b1 = np.array([[4, 1, 3],
               [5, 8, 5]])

print(f'{a1.shape}: a has {a1.shape[0]} rows and {a1.shape[1]} columns.')
print(f'{b1.shape}: b has {b1.shape[0]} rows and {b1.shape[1]} columns.')
print('Dimensions of result: (4x2)*(2x3)=(4x3)')
```

**Challenge**: Let's multiply `a1` with `b1`. Looking at the wikipedia example above, work out the values for c12 and c33 on paper. Then use the [`.matmul()`](https://numpy.org/doc/stable/reference/generated/numpy.matmul.html) function or the `@` operator to check your work. 

```python
# Matrix multiplication challenge: multiply a1 and b1
c = a1 @ b1
print('c = a1 @ b1 ->')
print(c)
print('c[0,1] =', c[0,1])
print('c[2,2] =', c[2,2])
```

```python
# show matmul using np.matmul as well
c2 = np.matmul(a1, b1)
print('c2 equals c ->', np.array_equal(c, c2))
print(c2)
```

# Manipulating Images as ndarrays


```python
# Manipulating images: load example image from scipy if available, else create a small RGB test image
try:
    img = misc.face()  # may exist in older scipy versions
except Exception:
    img = np.zeros((128,128,3), dtype=np.uint8)
    for i in range(128):
        img[i,:,0] = np.linspace(0,255,128)
        img[:,i,1] = np.linspace(0,255,128)
print('img dtype:', img.dtype)
print('img shape:', img.shape)
print('img ndim:', img.ndim)
plt.figure(figsize=(5,4))
plt.imshow(img)
plt.axis('off')
plt.title('Loaded image')
plt.show()
```

```python
# show shape/dtype/resolution and convert to sRGB floats
print('dtype:', img.dtype)
print('shape:', img.shape)
print('resolution (height x width):', img.shape[0], 'x', img.shape[1])
img_srgb = img.astype('float32') / 255.0 if img.dtype == np.uint8 else img.astype('float32')
print('converted to sRGB with dtype', img_srgb.dtype)
```
